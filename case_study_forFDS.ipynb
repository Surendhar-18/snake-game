```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student-cat.csv")
df

```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

df
```

```python
top_performers = df[df["Average"] >= 85][["Name", "Average"]]
top_performers
```

```python
support_students = df[df["Average"] < 50][["Name", "Average"]]
support_students
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Name"], df["Average"])
plt.xticks(rotation=45)
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.title("Student Performance Analysis")
plt.show()
```

```python
colors = []

for avg in df["Average"]:
    if avg >= 85:
        colors.append("green")   # Top performers
    elif avg < 50:
        colors.append("red")     # Needs support
    else:
        colors.append("blue")    # Average students

plt.figure(figsize=(10,5))
plt.bar(df["Name"], df["Average"], color=colors)
plt.xticks(rotation=45)
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.title("Student Performance Classification")
plt.show()
```
