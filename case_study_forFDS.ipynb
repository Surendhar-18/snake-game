# Student Performance Analysis - Case Study

This notebook analyzes student performance across multiple subjects.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load student data
df = pd.read_csv('student-cat.csv')
df
```

```python
# Calculate total and average scores
df['Total'] = df['Java'] + df['UIUX'] + df['OS'] + df['PQT'] + df['DataScience']
df['Average'] = df['Total'] / 5

df
```

```python
# Identify top performers (Average >= 85)
top_performers = df[df['Average'] >= 85][['Name', 'Average']]
print('Top Performers (Average >= 85):')
top_performers
```

```python
# Identify students needing support (Average < 50)
support_students = df[df['Average'] < 50][['Name', 'Average']]
print('Students Needing Support (Average < 50):')
support_students
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(df['Name'], df['Average'], color='skyblue', edgecolor='navy', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Student Name', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.title('Student Performance Analysis', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Create color-coded bar chart
colors = []
for avg in df['Average']:
    if avg >= 85:
        colors.append('green')
    elif avg < 50:
        colors.append('red')
    else:
        colors.append('blue')

plt.figure(figsize=(12, 6))
plt.bar(df['Name'], df['Average'], color=colors, edgecolor='black', alpha=0.8)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Student Name', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.title('Student Performance Analysis (Color-coded)', fontsize=14, fontweight='bold')
plt.axhline(y=85, color='green', linestyle='--', alpha=0.7, linewidth=2, label='Top Performer (85)')
plt.axhline(y=50, color='red', linestyle='--', alpha=0.7, linewidth=2, label='Support Threshold (50)')
plt.legend(loc='lower right', fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
